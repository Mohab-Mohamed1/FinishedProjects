# Tic Tac Toe Game 


## Game Description
This is an implementation of the popular game in which two players one with the "O" marker and the other wih the "X" marker play against each other to attempt to stack their markers horizontally or vertically along the board without any of the opponent's markers between them.

```python
"""
Created on Sun Oct 18 11:52:19 2020

@author: Mohab Mohamed
"""

'''
    Function Name : displayBoard
    Function description : Printing the board layout to the terminal
    Parameters : Board object 
    Return value : None
'''

def displayBoard(board):
    print("\t\t    __ __ __")
    print(f"\t\t    {board[0]} |{board[1]} |{board[2]}")
    print("\t\t    __ __ __")
    print(f"\t\t    {board[3]} |{board[4]} |{board[5]}")
    print("\t\t    __ __ __")
    print(f"\t\t    {board[6]} |{board[7]} |{board[8]}")
    

        
'''
    Function Name : PlayO
    Function description : Handles the "O" marker player move on the board
    Parameters : Board object 
    Return value : None
'''
def playO(board):
    while True:
        print("O player turn")
        try:
            play = int(input("Choose cell to play: "))
            play = play - 1
        except:
            print("Error! Enter a valid number between 1~9")
            continue
        
        if play < 0 or play > 8: #The input cell number is not between 0~8
            print("Error! Enter a play between 1~9")
            continue
        elif isinstance(board[play], str): #This cell already has a marker.
            print("Error! You can't play here")
            continue
        else:
            board[play] = 'O'
            break
            
'''
    Function Name : PlayX
    Function description : Handles the "X" marker player move on the board
    Parameters : Board object 
    Return value : None
'''          
def playX(board): # This function implement the X player move on the board
    while True:
        print("X player turn")
        try:
            play = int(input("Choose cell to play: "))
            play = play - 1
        except:
            print("Error! Enter a valid number between 1~9")
            continue
        if play < 0 or play > 8: #The input cell number is not between 0~8
            print("Error! Enter a play between 1~9")
            continue
        elif isinstance(board[play], str): #This cell already has a marker.
            print("Error! You can't play here")
            continue
        else:
            board[play] = 'X'
            break


'''
    Function Name : CheckForWin
    Function description : Check if winning condition is met
    Parameters : Board object 
    Return value : True if winning condition is met, False if winning condition is not met.
'''
def checkForWin(board): #check if a win condition was met
    if board[0] == board[4] and board[4] == board[8]:
        if board[0] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True    
    elif board[2] == board[4] and board[4] == board[6]:        
        if board[2] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True    
    elif board[0] == board[1] and board[1] == board[2]:
        if board[0] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True   
    elif board[3] == board[4] and board[4] == board[5]:
        if board[3] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True   
    elif board[2] == board[5] and board[5] == board[8]:
        if board[2] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True
    elif board[1] == board[4] and board[4] == board[7]:
        if board[1] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True   
    elif board[8] == board[7] and board[7] == board[6]:
        if board[8] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True
    elif board[6] == board[3] and board[3] == board[0]:    
        if board[6] == 'X':
            print("X player won")
        else:
            print("O player won")
        return True
    return False

'''
    Function Name : PlayerO
    Function description : Handles "O" marker player turn.
    Parameters : Board object 
    Return value : Boolean True or False
'''
def PlayerO(board):
    playO(board)
    displayBoard(board)
    return checkForWin(board)

'''
    Function Name : PlayerX
    Function description : Handles "O" marker player turn.
    Parameters : Board object 
    Return value : Boolean True or False
'''
def PlayerX(board):
    playX(board)
    displayBoard(board)
    return checkForWin(board)    
    

'''
    Function Name : GameEngine
    Function description : main function that runs the game
    Return value : None
'''
def GameEngine():
    print("\t\t________________")
    print("\t\tTic Tac Toe Game")
    print("\t\t________________")
    board = [1,2,3,4,5,6,7,8,9] # Board object 
    positions = 9
    displayBoard(board)
    if input("Start game? y/n : ") == 'y':
        while True:
            winO = PlayerO(board)
            positions = positions - 1
            if winO: #"O" player playes then check for win
                if input("Do you wanna play again?: ") == 'y':
                    board = [1,2,3,4,5,6,7,8,9]
                    positions = 9
                    displayBoard(board)
                    continue
                else:
                    break
            elif positions == 0:
                print("Draw")
                if input("Do you wanna play again?: ") == 'y':
                    board = [1,2,3,4,5,6,7,8,9]
                    positions = 9
                    displayBoard(board)
                    continue
                else:
                    break
                
            winX = PlayerX(board)
            positions = positions - 1
            if winX:
                if input("Do you wanna play again?: ") == 'y':
                    board = [1,2,3,4,5,6,7,8,9]
                    positions = 9
                    displayBoard(board)
                    continue
                else:
                    break
            elif positions == 0:
                print("Draw")
                if input("Do you wanna play again?: ") == 'y':
                    board = [1,2,3,4,5,6,7,8,9]
                    positions = 9
                    displayBoard(board)
                    continue
                else:
                    break
            
GameEngine()        
    
   
```
